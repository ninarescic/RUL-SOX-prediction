# LSTM Demo â€“ Time Series Regression (PyTorch)

This notebook demonstrates a **simple LSTM workflow using PyTorch** for time-series prediction.

We use **synthetic sine-wave data** to illustrate:
- sequence creation (windowing)
- time-aware train/test split
- an LSTM model in PyTorch
- training and evaluation

This notebook is a **robust Windows-friendly alternative** to TensorFlow/Keras.

## 1) Imports

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

print('Torch version:', torch.__version__)
```

## 2) Generate synthetic time series

We generate a noisy sine wave.
The task: predict the **next value** given a window of previous values.

```python
np.random.seed(42)
torch.manual_seed(42)

n_points = 1000
t = np.arange(n_points)
signal = np.sin(0.02 * t) + 0.1 * np.random.randn(n_points)

plt.figure(figsize=(8, 4))
plt.plot(signal)
plt.title('Synthetic time series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

## 3) Create sequences (windowing)

LSTMs expect input shape:
`(batch, sequence_length, features)`

```python
def create_sequences(series, window_size):
    X, y = [], []
    for i in range(len(series) - window_size):
        X.append(series[i:i+window_size])
        y.append(series[i+window_size])
    return np.array(X), np.array(y)

window_size = 30
X, y = create_sequences(signal, window_size)

X = X[..., np.newaxis]

print('X shape:', X.shape)
print('y shape:', y.shape)
```

## 4) Train / test split (time-aware)

We **do not shuffle** time series data.

```python
split = int(0.8 * len(X))

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)

print('Train:', X_train.shape)
print('Test :', X_test.shape)
```

## 5) Dataset and DataLoader

```python
class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_ds = TimeSeriesDataset(X_train, y_train)
test_ds = TimeSeriesDataset(X_test, y_test)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False)
```

## 6) Define LSTM model

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=32):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        return self.fc(out).squeeze()

model = LSTMModel()
print(model)
```

## 7) Training setup

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

## 8) Train the model

```python
n_epochs = 20
train_losses = []

for epoch in range(n_epochs):
    model.train()
    epoch_loss = 0.0

    for xb, yb in train_loader:
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    epoch_loss /= len(train_loader)
    train_losses.append(epoch_loss)

    print(f'Epoch {epoch+1}/{n_epochs} - Train MSE: {epoch_loss:.4f}')
```

### Training curve

```python
plt.figure(figsize=(6, 4))
plt.plot(train_losses)
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.title('Training loss')
plt.show()
```

## 9) Evaluate on test set

```python
model.eval()
with torch.no_grad():
    y_pred = []
    for xb, _ in test_loader:
        y_pred.append(model(xb))

y_pred = torch.cat(y_pred).numpy()

plt.figure(figsize=(8, 4))
plt.plot(y_test.numpy(), label='True')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.title('PyTorch LSTM prediction on test set')
plt.show()
```

## 10) Takeaways

- Sequence shape matters
- Time-aware splitting is essential
- This structure maps directly to RUL/SOX sequence models
